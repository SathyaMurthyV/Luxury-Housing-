```python
import sqlite3
import pandas as pd
conn = sqlite3.connect('my_local_db.sqlite')

# 1. Total row count

query1 = "SELECT COUNT(*) AS total_rows FROM Luxury_Housing_Bangalore;"
print(pd.read_sql(query1, conn))

# 2. Group by booking status

query2 = """
SELECT Possession_Status, COUNT(*) AS total
FROM Luxury_Housing_Bangalore
GROUP BY Possession_Status;
"""
print(pd.read_sql(query2, conn))

# 3 . Average ticket price per builder

query3 = """
SELECT Developer_Name, AVG(Ticket_Price_Cr) AS avg_price
FROM Luxury_Housing_Bangalore
GROUP BY Developer_Name
ORDER BY avg_price DESC;
"""

print(pd.read_sql(query3, conn))
conn.close()
```
